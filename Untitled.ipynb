```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from plotly.offline import plot
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns

import pickle
from wordcloud import WordCloud

```

# Load and store dataset

```python
df = pd.read_csv('amazon.csv')
df.head(7)
```

# Pre-process the text data

```python
def clean_text(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # Remove punctuation
    text = text.lower()
    words = text.split()
    words = [PorterStemmer().stem(w) for w in words if w not in stopwords.words('english')]
    return " ".join(words)

df['Clean_Review'] = df['review_content'].astype(str).apply(clean_text)
```

```python
(df.columns)
```

# Sentiment Labeling 

```python
# Clean the rating column
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # '|' or "abc" or other thing → NaN
df['rating'] = df['rating'].fillna(0)  # or dropna() use

#  Apply sentiment labeling
def label(rating):
    if rating >= 4:
        return 'positive'
    elif rating <= 2:
        return 'negative'
    else:
        return 'neutral'

df['Sentiment'] = df['rating'].apply(label)
```

# Vectorization + Model Training

```python
tfidf = TfidfVectorizer(max_features=3000)
X = tfidf.fit_transform(df['Clean_Review'])
y = df['Sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = LogisticRegression()
model.fit(X_train, y_train)
print(classification_report(y_test, model.predict(X_test)))
```

# wordcloud in differnt shape

```python
# Dummy Data Preparation (if df is not already loaded)
data = {
    'Sentiment': np.random.choice(['Positive', 'Neutral', 'Negative'], size=100),
    'Clean_Review': np.random.choice([
        "Love this product", "Very useful", "Excellent quality", "Not good", "Worst experience", "Okay type",
        "Highly recommended", "Bad service", "Average", "Superb"
    ], size=100)
}
df = pd.DataFrame(data)

#  Make Sentiment column lowercase to avoid case mismatch
df['Sentiment'] = df['Sentiment'].str.lower()

# Safe check for positive reviews
positive_reviews = df[df['Sentiment'] == 'positive']['Clean_Review']

if positive_reviews.empty:
    print(" No positive reviews found. Using fallback dummy text.")
    positive_text = "positive review placeholder"
else:
    positive_text = " ".join(positive_reviews.astype(str))

# square:
wc_square = WordCloud(width=400, height=400, background_color='white').generate(positive_text)

#  Circle:
x, y = np.ogrid[:400, :400]
mask_circle = (x - 200)**2 + (y - 200)**2 > 180**2
mask_circle = 255 * mask_circle.astype(int)
wc_circle = WordCloud(width=400, height=400, background_color='white', mask=mask_circle).generate(positive_text)

#  Heart shape:
def heart_mask(width=400, height=400):
    x = np.linspace(-1.5, 1.5, width)
    y = np.linspace(-1.5, 1.5, height)
    x, y = np.meshgrid(x, y)
    mask = ((x**2 + y**2 - 1)**3 - x**2 * y**3) <= 0
    mask = np.flipud(mask)
    return np.where(mask, 0, 255).astype(np.uint8)

mask_heart = heart_mask()
wc_heart = WordCloud(width=400, height=400, background_color='white',
                     mask=mask_heart, contour_color='red', contour_width=2,
                     colormap='Reds').generate(positive_text)

# Bulb shape:
def bulb_mask(width=400, height=400):
    x = np.linspace(-1, 1, width)
    y = np.linspace(-1.5, 1.5, height)
    x, y = np.meshgrid(x, y)
    top = ((x / 0.6) ** 2 + (y / 0.9) ** 2) <= 1
    bottom = (np.abs(x) < 0.3) & (y > 0.7) & (y < 1.3)
    mask = top | bottom
    return np.where(mask, 0, 255).astype(np.uint8)

mask_bulb = bulb_mask()
wc_bulb = WordCloud(width=400, height=400, background_color='white',
                    mask=mask_bulb, contour_color='orange', contour_width=2,
                    colormap='autumn').generate(positive_text)

#  Mango shape:
def mango_mask(width=400, height=400):
    x = np.linspace(-1.5, 1.5, width)
    y = np.linspace(-1.5, 1.5, height)
    x, y = np.meshgrid(x, y)
    body = ((x) ** 2 + (y / 1.2) ** 2) < 1
    bite = ((x - 0.5)**2 + (y - 0.3)**2) < 0.3
    stem = (np.abs(x) < 0.1) & (y < -0.9) & (y > -1.3)
    mask = (body & ~bite) | stem
    return np.where(mask, 0, 255).astype(np.uint8)

mask_mango = mango_mask()
wc_mango = WordCloud(width=400, height=400, background_color='white',
                     mask=mask_mango, contour_color='darkred', contour_width=2,
                     colormap='YlOrRd').generate(positive_text)

# Cloud
def custom_cloud_mask(width=400, height=400):
    mask = np.ones((height, width), dtype=np.uint8) * 255
    def draw_circle(mask, cx, cy, r):
        y, x = np.ogrid[:mask.shape[0], :mask.shape[1]]
        area = (x - cx) ** 2 + (y - cy) ** 2 <= r ** 2
        mask[area] = 0
    draw_circle(mask, 130, 160, 60)
    draw_circle(mask, 200, 140, 70)
    draw_circle(mask, 270, 160, 60)
    draw_circle(mask, 170, 220, 75)
    draw_circle(mask, 240, 230, 65)
    return mask

mask_cloud = custom_cloud_mask()
wc_cloud = WordCloud(width=400, height=400, background_color='white',
                     mask=mask_cloud, contour_color='skyblue', contour_width=2,
                     colormap='Blues').generate(positive_text)

# Plot All WordClouds ---
fig, axs = plt.subplots(2, 3, figsize=(16, 10))
axs = axs.flatten()

titles = ["Square", "Circle", "Heart", "Bulb", "Mango", "Cloud"]
clouds = [wc_square, wc_circle, wc_heart, wc_bulb, wc_mango, wc_cloud]

for i in range(6):
    axs[i].imshow(clouds[i], interpolation='bilinear')
    axs[i].axis('off')
    axs[i].set_title(titles[i])

plt.tight_layout()
plt.show()
```

# Kmeans clustraing

```python
#i use kmeans clutraning...it's compare rating and reviews_length
```

```python
# Create Review_Length feature
df['Review_Length'] = df['Clean_Review'].apply(lambda x: len(str(x).split()))

# Select features for clustering
X_cluster = df[['rating', 'Review_Length']]

# Scale features for better clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# Run KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Show first few rows with clusters
print(df[['rating', 'Review_Length', 'Cluster']].head(10))

# Show cluster-wise average rating and review length
print("\nCluster group means:")
print(df.groupby('Cluster')[['rating', 'Review_Length']].mean())

# Plot the clusters
plt.figure(figsize=(8,6))
scatter = plt.scatter(df['rating'], df['Review_Length'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Rating')
plt.ylabel('Review Length')
plt.title('KMeans Clusters of Reviews')
plt.colorbar(scatter, label='Cluster')
plt.show()
```

```python
from transformers import pipeline

# Summarizer load
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

text = """
Bangladesh is a riverine country, crisscrossed by numerous rivers that are vital to its geography, economy, and culture.
The Padma, Jamuna, and Meghna are the major rivers, originating from the Himalayas and eventually emptying into the Bay of Bengal.
These rivers, along with their tributaries, form a vast network that supports agriculture, transportation, and fishing industries,
making them essential to the nation's livelihood."""

summary = summarizer(text, max_length=100, min_length=30, do_sample=False)
print("Summary:\n", summary[0]['summary_text'])
```
